```python

import math

#!Hàm có nhiệm vụ làm tròn các số có giá trị gần bằng 0 
#! và làm tròn các kết quả khác tới số thập phân thứ 4
def round_num(value):
    if(math.isclose(value,0) == True):
        value = 0
    else:
        value = round(value,4) 
    return value

#! hàm nhân ma trận
def multiMatrix(a,b): #! m.n & n.p = m.p
    m = len(a) #! số dòng của a
    n = len(b[0]) #! số cột của b
    #!số cột a phải bằng số dòng b
    if len(a[0]) != len(b):
        print("khong the nhan 2 ma tran")
        return None
    else:
        #! tạo ma trận m.n 
        c = [[0 for _ in range(n)] for _ in range(m)]
    #! Lấy từng cột của a nhân với từng dòng của b 
    for i in range(m):
        for j in range(n):
            total = 0
            for k in range (len(a[0])):
                total += a[i][k] * b[k][j]
            c[i][j] = round_num(total)
            #c[i][j] = total
    return c

#! hàm tạo ma trận đơn vị
def matran_donvi(a):
    c = [[0 for _ in range(len(a))] for _ in range(len(a))]
    for i in range(len(a)):
        c[i][i] = 1
    return c

#! hàm tính lũy thừa ma trận
def luy_thua(a,k):
    #! kiểm tra ma trận vuông
    if len(a) != len(a[0]):
        print("Khong phai la ma tran vuong")
        return None
    else:
        a_k = matran_donvi(a)
        i = 1
        while i <= k:
            a_k = multiMatrix(a_k,a)
            i+= 1
        return(a_k)

#! hàm tính phân phối
def tinh_phan_phoi(P_matrix,Pi,n):
    if n == 0:
        return(Pi)
    else:
        res = luy_thua(P_matrix,n)
        if res == None:
            print("Khong tinh duoc luy thua ma tran P")
            return None
        else:
            c = multiMatrix(res,Pi)
            return(c)
            
def compare_matrix(a,b):
    if len(a) != len(b) or len(a[0]) != len(b[0]):
        return False
    else:
        check = False
        for i in range(len(a)):
            for j in range(len(a[0])):
                if a[i][j] != b[i][j]:
                    return False
        return True

def get_col(a):
    res = []
    for i in range(len(a)):
        res.append([a[i][0]])
    return res

def tinh_phan_phoi_dung(a):
    stop = False
    i = 1
    while stop == False:
        ex = luy_thua(a,i)
        col = get_col(ex)
        print(col)
        res = multiMatrix(ex,col)
        print(res)
        print("\n")
        if compare_matrix(res,col) == True:
            return col
        i += 1


#! main
#a = [[1/3,3/4],[2/3,1/4]]
a = [[1/3,1/2,0],[1/3,0,1/4],[1/3,1/2,3/4]]
b = [[0,1578],[0,2105],[0,6317]]
print(b)
print(multiMatrix(a,b))


```
