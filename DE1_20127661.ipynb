```python
from fractions import Fraction #! thư viện để lấy phân sô từ input
import math #! thư viện để làm tròn kết quả tính toán

#! MSSV: 20127661
#! Họ và Tên: Lê Phan Duy Tùng
#! Lớp 20CLC01


#! hàm thực hiện câu 1
def cau1():
    #! Lấy số trạng thái
    m = int(input("nhap so trang thai: "))
    #! Tạo tập trạng thái
    S = []
    for i in range (1,m+1):
        S.append(i)
    #! Phân phối đầu
    Pi = []
    for i in range (m):
        temp = input("Nhap phan tu thu "+ str(i+1)+": ")
        temp = Fraction(temp)
        temp = float(temp)
        Pi.append([temp])
    #! Tạo ma trận chuyển 
    P = [[0 for _ in range(m)] for _ in range(m)]
    for i in range (m):
        for j in range (m):
            temp = input("Nhap phan tu thu a["+ str(i)+"]["+str(j)+"]: ")
            temp = Fraction(temp)
            temp = float(temp)
            P[i][j] = temp
    return S,Pi,P

#!Hàm có nhiệm vụ làm tròn các số có giá trị gần bằng 0 
#! và làm tròn các kết quả khác tới số thập phân thứ 4
def round_num(value):
    if(math.isclose(value,0) == True):
        value = 0
    else:
        value = round(value,4) 
    return value

#! hàm tạo ma trận đơn vị
def matran_donvi(a):
    c = [[0 for _ in range(len(a))] for _ in range(len(a))]
    for i in range(len(a)):
        c[i][i] = 1
    return c

#! hàm nhân ma trận
def multiMatrix(a,b): #! m.n & n.p = m.p
    m = len(a) #! số dòng của a
    n = len(b[0]) #! số cột của b
    #!số cột a phải bằng số dòng b
    if len(a[0]) != len(b):
        print("khong the nhan 2 ma tran")
        return None
    else:
        #! tạo ma trận m.n 
        c = [[0 for _ in range(n)] for _ in range(m)]
    #! Lấy từng cột của a nhân với từng dòng của b 
    for i in range(m):
        for j in range(n):
            total = 0
            for k in range (len(a[0])):
                total += a[i][k] * b[k][j]
            c[i][j] = round_num(total)
            #c[i][j] = total
    return c

#! hàm tính lũy thừa ma trận
def luy_thua(a,k):
    #! kiểm tra ma trận vuông
    if len(a) != len(a[0]):
        print("Khong phai la ma tran vuong")
        return None
    else:
        a_k = matran_donvi(a)
        i = 1
        while i <= k:
            a_k = multiMatrix(a_k,a)
            i+= 1
        return(a_k)

#! hàm tính phân phối xác xuất
def tinh_phan_phoi(P_matrix,Pi,n):
    if n == 0:
        return(Pi)
    else:
        res = luy_thua(P_matrix,n)
        if res == None:
            print("Khong tinh duoc luy thua ma tran P")
            return None
        else:
            c = multiMatrix(res,Pi)
            return(c)

#! hàm thực hiện câu 2
def cau2():
    res = cau1()
    P_matrix = res[2]
    Pi = res[1]
    n = int(input("nhap gia tri n: "))
    c = tinh_phan_phoi(P_matrix, Pi, n)  
    return c

#! hàm kiểm tra 2 ma trận có bằng nhau
def compare_matrix(a,b):
    if len(a) != len(b) or len(a[0]) != len(b[0]):
        return False
    else:
        check = False
        for i in range(len(a)):
            for j in range(len(a[0])):
                if a[i][j] != b[i][j]:
                    return False
        return True

#! hàm lấy cột dầu tiên từ 1 ma trận
def get_col(a):
    res = []
    for i in range(len(a)):
        res.append([a[i][0]])
    return res

#! hàm tính phân phối dừng
def tinh_phan_phoi_dung(a):
    i = 1    
    ex1 = luy_thua(a,100)
    ex2 = luy_thua(a,101)
    ex3 = luy_thua(a,103)
    if compare_matrix(ex1,ex2) == True and compare_matrix(ex2,ex3) == True:
        col_matrix = get_col(ex1)
        return col_matrix
    else:
        print("khong tim duoc phan phoi dung")
        return None
def cau3():
    res = cau1()
    P_matrix = res[2]
    return tinh_phan_phoi_dung(P_matrix) 
    

#! main
#! câu 4.a
# print("cau 1")
# res1 =  cau1()
# print("Tap trang thai: ")
# print(res1[0])
# print("Phan phoi dau: ")
# print(res1[1])
# print("Ma tran chuyen da nhap: ")
# print(res1[2])
#! câu 4.b
# print("cau 2")
# print("Phan phoi tai thoi diem n = 4 la: ")
# print(cau2())
#! câu 4.c
print("cau 3")
print("Phan phoi dung la: ")
print(cau3())
```
